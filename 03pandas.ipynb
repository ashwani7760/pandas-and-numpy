```python
import pandas as pd
```

```python
d1  = {'key1' : [1,2,3,4,5],
      "key2" : [6,7,8,9,10],
      "key3" : [11,12,13,"ash",15]}
```

```python
data1 = pd.DataFrame(d1)
```

```python
d2  = {'key1' : [1,2,"ap",4,5],
      "key2" : [6,7,"sd",9,10],
      "key3" : [11,"as",13,"ash",15]}
```

```python
data2 = pd.DataFrame(d2)
```

```python
print(data1)
print(data2)
```

```python
pd.merge(data1,data2)  # by default how = inner     inner like intersection in set       data1 consider as left and other right
```

```python
pd.merge(data1,data2,how = "inner",on = "key1")  
```

```python
pd.merge(data1,data2,how = "left",on = "key1")  
```

```python
pd.merge(data1,data2,how = "right",on = "key1")  
```

```python
pd.merge(data1,data2,how = "outer",on = "key1")  
```

```python
d1  = {'a' : [1,2,3,4,5],
      "b" : [6,7,8,9,10],
      "c" : [11,12,13,"ash",15]}

d2  = {'key1' : [1,2,"ap",4,5],
      "key2" : [6,7,"sd",9,10],
      "key3" : [11,"as",13,"ash",15]}

df11 = pd.DataFrame(d1)
df12 = pd.DataFrame(d2)
```

# how to join when not any column name is common in both dataframe

```python
pd.merge(df11,df12,how="inner",on= "key1")
```

```python
pd.merge(df11,df12,how="inner",left_on= "a", right_on = "key1")
```
